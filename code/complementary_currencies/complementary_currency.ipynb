```python
entries = []
for page in range(0,11):
    driver = webdriver.Chrome()
    driver.get("http://complementarycurrency.org/ccDatabase/les_public_splash.php?action=list&which=" + str(page) + "&list=r")
    html = driver.page_source
    doc = BeautifulSoup(html, 'html.parser')
    my_table = doc.find("table", attrs={"width": "100%"})
    each_entry = my_table.find_all('tr')[4:34]
    for entry in each_entry:
        ENTRY_DICT = {}
        entry_name = entry.find_all('ul')[0].find('a')
        if entry_name:
            #print(entry_name.text.strip())
            ENTRY_DICT['system_name'] = entry_name.text.strip()
        entry_type = entry.find_all('ul')[0].find('li')
        if entry_type:
            #print(entry_type.text.strip())
            ENTRY_DICT['system_type'] = entry_type.text.strip()
        entry_origin = entry.find_all('td')[2]
        if entry_origin:
            #print(entry_origin.text.strip())
            ENTRY_DICT['system_origin'] = entry_origin.text.strip()
        entry_serving = entry.find_all('td')[1]
        if entry_serving:
            #print(entry_serving.text.strip())
            ENTRY_DICT['system_serving'] = entry_serving.text.strip()
        entry_year_of_establishment = entry.find_all('td')[3]
        if entry_year_of_establishment:
            #print(entry_year_of_establishment.text.strip())
            ENTRY_DICT['establishment'] = entry_year_of_establishment.text.strip()
        entries.append(ENTRY_DICT)
        print(ENTRY_DICT)
        print("-----")
```

```python
df = pd.DataFrame(entries)
df
```

```python
df['establishment'] = df['establishment'].str.extract(r"(\d\d\d\d)")
```

```python
df['system_type'] = df['system_type'].str.extract(r"\n\n(.*)")
```

```python
df.to_csv("currencies.csv", index=False)
```
