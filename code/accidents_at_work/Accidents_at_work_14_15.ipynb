```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
```

# 2014

```python
sector_region_14 = pd.read_excel("economic_activity_14.xlsx")
sector_region_14
```

```python
sector_region_14.groupby('sector')['number_of_accidents14'].sum().sort_values().plot(kind='barh', figsize=(10,10))
```

# 2015

```python
sector_region_15 = pd.read_excel("economic_activity_15.xlsx")
sector_region_15
```

```python
sector_region_15.groupby('sector')['number_of_accidents15'].sum().sort_values().plot(kind='barh', figsize=(10,10))
```

```python
combined_df = sector_region_14.merge(sector_region_15, how='left', on='sector')
combined_df.head()
```

```python
combined_df.set_index('sector', inplace=True)
combined_df
```

```python
combined_df.plot(kind='barh', figsize=(10,5))

plt.savefig('sector_total_14_15.pdf')
```
