```python
%matplotlib inline
import pandas as pd
from selenium import webdriver
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
driver = webdriver.Chrome()
driver.get('http://www.whosampled.com/The-Winstons/Amen,-Brother/sampled/')
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
tracks_all = []

for page in range(1,167):
    driver.get("http://www.whosampled.com/The-Winstons/Amen,-Brother/sampled/?cp=" + str(page))
    doc = BeautifulSoup(driver.page_source, 'html.parser')
    each_track = doc.find_all(class_="listEntry sampleEntry")[0:16]
    for track in each_track:
        all_track_dict ={}
        track_title = track.find_all('a')[1]
        track_artist = track.find_all('a')[2]
        track_genre = track.find('span', class_="bottomItem")
        track_year = track.find(class_="trackArtist")
#         print('Title:', track_title.text.strip())
#         print('Artist:', track_artist.text.strip())
#         print('Genre:', track_genre.text.strip())
#         print('Year:', track_year.text.strip())
        all_track_dict['Title'] = track_title.text.strip()
        all_track_dict['Artist'] = track_artist.text.strip()
        all_track_dict['Genre'] = track_genre.text.strip()
        all_track_dict['Year'] = track_year.text.strip()
        tracks_all.append(all_track_dict)
        print(all_track_dict)
        print('-----')
```

```python
df = pd.DataFrame(tracks_all)
df
```

```python
df['Year'] = df['Year'].str.extract(r"(\d\d\d\d)")
```
